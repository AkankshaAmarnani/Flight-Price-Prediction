```python
airline_bm=[]
date_of_journey_bm=[]
month_of_journey_bm=[]
source_bm=[]
destination_bm=[]
dep_time_bm=[]
arrival_time_bm=[]
duration_bm=[]
total_stops_bm=[]
price_bm=[]

time.sleep(5)
#Lets extract all the tags having the airline
airline_tags=driver.find_elements_by_xpath("//div[@class='col-2 col-md-2']/span[2]") #locating web element of airline
for i in airline_tags:
    if i.text is None:
        airline_bm.append("NA")
    else:
        v=i.text
        airline_bm.append(v)
        
#Lets extract all the tags having the date_of_journey
date_of_journey_tags=driver.find_elements_by_xpath("//div[@class='col-2 col-md-2']/span[2]") #locating web element of date_of_journey
for i in date_of_journey_tags:
    if i.text is None:
        date_of_journey_bm.append("NA")
    else:
        r=i.text
        date_of_journey_bm.append(29)
        
#Lets extract all the tags having the month_of_journey
month_of_journey_tags=driver.find_elements_by_xpath("//div[@class='col-2 col-md-2']/span[2]") #locating web element of month_of_journey
for i in month_of_journey_tags:
    if i.text is None:
        month_of_journey_bm.append("NA")
    else:
        v=i.text
        month_of_journey_bm.append("December")
        
#Lets extract all the tags having the source
source_tags=driver.find_elements_by_xpath("//div[@class='col-2 col-md-2']/span[2]") #locating web element of source_dm
for i in source_tags:
    if i.text is None:
        source_bm.append("NA")
    else:
        s=i.text
        source_bm.append("Banglore")
            
#Lets extract all the tags having the destination
destination_tags=driver.find_elements_by_xpath("//div[@class='col-2 col-md-2']/span[2]") #locating web element of destination
for i in destination_tags:
    if i.text is None:
        destination_bm.append("NA")
    else:
        d=i.text
        destination_bm.append("Mumbai")
        
#Lets extract all the tags having the dep_time
dep_time_tags=driver.find_elements_by_xpath("//span[@class='departure-time base-time']") #locating web element of dep_time
for i in dep_time_tags:
    if i.text is None:
        dep_time_bm.append("NA")
    else:
        dep=i.text
        dep_time_bm.append(dep)
            
#Lets extract all the tags having the arrival_time
arrival_time_tags=driver.find_elements_by_xpath("//span[@class='arrival-time base-time']") #locating web element of arrival_time
for i in arrival_time_tags:
    if i.text is None:
        arrival_time_bm.append("NA")
    else:
        arr=i.text
        arrival_time_bm.append(arr)           

#Lets extract all the tags having the duration
duration_tags=driver.find_elements_by_xpath("//div[@class='col-field stops ml-auto mr-auto']/div[1]") #locating web element of duration
for i in duration_tags:
    if i.text is None:
        duration_bm.append("NA")
    else:
        dur=i.text
        duration_bm.append(dur)            
            
#Lets extract all the tags having the total_stops
total_stops_tags=driver.find_elements_by_xpath("//div[@class='col-field stops ml-auto mr-auto']/div[3]") #locating web element of total_stops
for i in total_stops_tags:
    if i.text is None:
        total_stops_bm.append("NA")
    else:
        ts=i.text
        total_stops_bm.append(ts)      
            
#Lets extract all the tags having the price
price_tags=driver.find_elements_by_xpath("//span[@class='money d-none d-sm-block']") #locating web element of price
for i in price_tags:
    if i.text is None:
        price_bm.append("NA")
    else:
        pr=i.text
        price_bm.append(pr.replace('',''))                       
  
pd.set_option("display.max_rows", None, "display.max_columns", None)                #To display the entire dataframe
df_bm=pd.DataFrame({'Airline':airline_bm, 'Date of Journey':date_of_journey_bm, 'Month of Journey': month_of_journey_bm, 'Source': source_bm, 'Destination': destination_bm, 'Departure Time' :dep_time_bm, 'Arrival Time': arrival_time_bm, 'Duration':duration_bm, 'Total Stops':total_stops_bm, 'Price': price_bm })
df_bm
```
